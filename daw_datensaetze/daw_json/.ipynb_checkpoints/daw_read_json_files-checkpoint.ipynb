```python
import pandas as pd
import numpy as np
```

```python
# Json File einlesen
df_raw = pd.read_json("https://raw.githubusercontent.com/ankitgoel1602/data-science/master/json-data/multiple_levels.json")
df_raw.head()
```

```python
df_nested = pd.json_normalize(df_raw["Results"])
df_nested
```

```python
df1 = pd.json_normalize(df_nested[0])
df1
```

```python
df2 = pd.json_normalize(df_nested[1])
df2
```

```python
df3 = pd.json_normalize(df_nested[2])
df3
```

```python
df4 = pd.json_normalize(df_nested[3])
df4
```

```python
df_all = [df1, df2, df3, df4]
df_unnested = pd.concat(df_all, axis = 0).reset_index(drop=True)
df_unnested
```

```python
import json
```

```python
with open("multiple_levels.json",'r') as f:
    data = json.loads(f.read())
```

```python
df_multiple_levels = pd.json_normalize(data, 
                                       record_path=["Results"], 
                                       meta = ["original_number_of_clusters", 
                                               "family_min_samples_percentage",
                                               "Scaler"], 
                                       record_prefix='dbscan_')
df_multiple_levels
```
