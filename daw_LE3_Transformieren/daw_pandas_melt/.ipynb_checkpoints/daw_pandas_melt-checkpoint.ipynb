```python
import pandas as pd
import numpy as np
```

```python
# Dataframe erstellen
data = {'Name': ["Ben","Julia","Gabu"], '08/08/21': [10,20,30], '10/08/21': [40,140,32], '12/08/21': [54,23,64]}
df = pd.DataFrame(data)
df
```

```python
# Melting mit id_vars

df.melt(id_vars = "Name")

# Wahle die Spalte aus, die sich nicht aendern sollte.
# Statt nun Zeile und Spalte zu betrachten, kann ich nun anhand einer Zeile nun den Werte ablesen.
# Ben mit der Variabel 08/08/21 hat en Betrag 10. 

# Vorher:
# Zeile Ben, Spalte 08/08/21, dort finde ich den Wert 10. 
```

```python
# Melting nach 2 Spalten
df.melt(id_vars = "Name", value_vars=["08/08/21","10/08/21"])

# Selektieren von nur den gewuenschten Spalten. 
```

```python
# Melting und Variabel und Value Spalte umbennnen:
df.melt(id_vars = "Name", value_vars=["08/08/21","10/08/21"], var_name="Datum", value_name="Kosten")
```
