```python
import pandas as pd
import numpy as np
```

```python
titanic = pd.read_csv("test.csv")
titanic.head()
```

```python
# Spaltenwerte umwandeln
```

```python
titanic["Embarked"] = titanic.Embarked.factorize()[0]
titanic.head()
```

```python
# Quantiative Werte zu QUalitative umwandeln
```

```python
titanic["Age_Kat"] = pd.cut(titanic.Age, bins = [0, 18, 25, 99], labels = ["child", "young adult", "adult"])
```

```python
titanic.head()
```

```python
titanic["Age_Kat_nums_right_true"] = pd.cut(titanic.Age, bins = [0, 18, 25, 99], right = True)
```

```python
titanic["Age_Kat_nums_right_false"] = pd.cut(titanic.Age, bins = [0, 18, 25, 99], right = False)
```

```python
titanic.head()
```

```python
# using qcut
titanic["Fare_cat"] = pd.qcut(titanic.Fare, q=3)
titanic["Fare_cat_label"] = pd.qcut(titanic.Fare, q=3, labels = ["low","medium","high"])
```

```python
titanic.head()
```
