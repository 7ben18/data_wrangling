```python
# Modul importieren
import pandas as pd
```

```python
# Datensatz einlesen
df = pd.read_excel("homelessness.xlsx", header=None)
df.head(2)
```

```python
df.shape
```

```python
# Spalte ansprechen und werte splitten und als Tabelle darstellen. 
df = df[0].str.split(",", expand = True)
df.head(2)
```

```python
# Eine Zeile als Kopfzeile definieren
header = list(df.iloc[0])
df.columns = header
df.head(2)
```

```python
# Die erste Zeile dann entfernen, weil diese als Kopfzeile dient
df = df.drop(0, axis = 0)
df.head(2)
```

```python
# Die erste Spalte entfernen, da nicht noetig
df = df.drop("", axis = 1)
df.head(2)
```

```python
# Datafarme nach region sortieren
df = df.sort_values("region", ascending = True)
df.reset_index(drop=True)
```

```python
df.shape
```

```python
# In der Spalte Region interessieren mich alle ausser Mid Atlantic
df2 = df[df["region"] != "Mid-Atlantic"]
df2.shape
```

```python
# In der Spalte Region, interessieren mich nur Mid Atlantic
df_tilde = df[~(df["region"] != "Mid-Atlantic")]
df_tilde
```

```python
# In der Region, Spalte interessieren mich alle, ausser Mountain und South Atlantic
df3 = df[(df["region"] != "Mountain") & (df["region"] != "South Atlantic")]
df3
```

```python
# In der Region Spalte interessieren mich nur Mountain und South Atlantic, ausser Mountain und South Atlantic
df[(df["region"] == "Mountain") | (df["region"] == "South Atlantic")]
```

```python
# Checken, ob der Wert in der Spalte vorhanden ist.
liste = ["Pacific", "New England", "South Atlantic"]
df_logic = df["region"].isin(liste)
df[df_logic]
```

```python
# Alle Werte ausgeben, die der Liste nicht entsprechen
liste = ["Pacific", "New England", "South Atlantic"]
df_logic2 = ~(df["region"].isin(liste))
df[df_logic2]
```
