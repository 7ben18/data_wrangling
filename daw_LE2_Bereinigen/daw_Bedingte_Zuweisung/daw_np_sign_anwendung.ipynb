```python
import pandas as pd
import numpy as np
```

```python
gameplay = {"Team1_P": np.random.randint(10, size = 100000),
            "Team2_P": np.random.randint(10, size = 100000)}
game_df = pd.DataFrame(gameplay)
game_df
```

```python
# Calculating difference
game_df["T1_difference"] = game_df["Team1_P"] - game_df["Team2_P"]


game_df["win_tie_lose_for_T1"] = np.sign(game_df["T1_difference"])
game_df

```
