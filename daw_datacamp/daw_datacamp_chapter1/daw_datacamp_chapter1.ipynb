# Spalte hinzufuegen (Mutieren/Transformieren)

```python
data2 = pd.read_csv("homelessness.csv", index_col="Unnamed: 0")
```

```python
data2.head(7)
```

# Spalte hinzufuegen

```python
# Spalte hinzufuegen
data2["individuals/state_pop"] = data2["individuals"] / data2["state_pop"]

# Spalte wird gefiltert
data_filter = data2[data2["individuals/state_pop"] >= 0.001014]

# Spalte wird sortiert, Absteigend
data_filter = data_filter.sort_values("individuals/state_pop", ascending=False)

# Vom DF nur die gewuenschten selektieren
data_filter[["individuals","state_pop","individuals/state_pop"]].head(10)
```

```python
data2.sort_values("individuals/state_pop")
```

```python
# Einlesen der Daten und Zuweisung einer Variabel
homelessness = pd.read_csv("homelessness.csv", index_col="Unnamed: 0")

# Zusaetliche Spalte hinzugefuegt
homelessness["indiv_per_10k"] = 10000 * homelessness["individuals"] / homelessness["state_pop"]
homelessness.head()

# Datensatz filtern
high_homelessness = homelessness[homelessness["indiv_per_10k"] > 20]

# Datensatz sortieren nach indiv_per_10k
high_homelessness_srt = high_homelessness.sort_values("indiv_per_10k", ascending = False)

# Vom Datensatz nur die gewuenschten Spalten auswaehlen
high_homelessness_srt[["indiv_per_10k", "state"]]
```
